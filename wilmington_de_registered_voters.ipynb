# Merging Two DataFrames

Using the data from the documentation, create a Pandas DataFrame that contains the party code and the corresponding legal name for each party.

```python
legal_names = {'A':'American','B':'American Delta','D':'Democratic Party','E':'Liberal','G':'Non Partisan','H':'Green Party','I':'Unaffiliated','J':'Working Families Party','L':'Libertarin','M':'Reform Party','O' :'Others/All Other Parties','Q': 'Constitution Party','R': 'Republican Party','T':'Conservative','V':'Natual Law Party','W': 'Independent Party of Del' ,'Y': 'The Blue Enigma Party','Z':'Socialist Workers Party'}
```

```python
df_legal_names = pd.DataFrame(legal_names.items(), columns = ['Party Code','Party Name'])

```

```python
df_legal_names
```

Create a Pandas DataFrame that contains the Party code and the corresponding number of registered voters for said party.

```python
# Create DataFrame df_party_counts here 
df_party_counts = pd.DataFrame(grouped_p.size(), columns = ['Count'])
df_party_counts.reset_index(inplace=True)
```

```python
df_party_counts 
```

Merge the two DataFrames.

```python
# Merge DataFrame here 
frames = [df_legal_names,df_party_counts ]
df = pd.concat(frames, axis=1)
df1 = df.drop(columns='PARTY', axis=1)
```

```python
# Display new DataFrame here 
df1
```

Create a bar graph again but using the legal name for each party as the value on the x axes.

```python
df1.plot.bar(y="Count", x='Party Name', title="Voters by Party")
```
